```python
def convert_people_cell(cell):
    if cell=='n.a.':
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell=='not available':
        return None
    return cell

df = pd.read_excel('/content/drive/MyDrive/008-Interviews/Pandas_Projects/Data/stock_data.xlsx', 'Sheet1', converters = {
    'people': convert_people_cell,
    'eps': convert_eps_cell
})
df
```

```python
df.to_excel('/content/drive/MyDrive/008-Interviews/Pandas_Projects/Data/new.xlsx', index=False , sheet_name='stocks', startrow=1, startcol=2)
```
