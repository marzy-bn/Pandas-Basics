```python
import pandas as pd
import numpy as np
#create data
stuff = {
    'Corporation':['Apple','Google','Meta','Apple','Google','Meta'],
    'Employees':['John','April','Wes','Beth','Justin','Steph'],
    'Salary':[200, 220, 190, 130, 120, 150]}
#create dataframe
my_df = pd.DataFrame(stuff)
my_df
```

```python
#Group by Corporation - to get object location in memory
company = my_df.groupby('Corporation')
company
```

```python
#sum
company.sum()
```

```python
#mean
company.mean()
```

```python
#max min
company.max()
```

```python
#standard deviation
company.std()
```

```python
#variance
company.var()
```

```python
#count
company.count()
```

```python
#describe
company.describe()
```
